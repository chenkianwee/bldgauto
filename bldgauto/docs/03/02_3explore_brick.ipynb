# Explore a brick file using Python rdflib built in functions

```python
from pathlib import Path
from rdflib import Graph, RDF, Namespace

ttl_filepath = Path.cwd().joinpath('brick_bsll.ttl')
BRICK = Namespace('https://brickschema.org/schema/Brick#')
REF = Namespace("https://brickschema.org/schema/Brick/ref#")
MODBUS = Namespace("https://www.modbus.org/")
bldg_graph = Graph()
bldg_graph.parse(ttl_filepath)
```

```python
for s, p, o in bldg_graph.triples((None, RDF.type, BRICK.Humidity_Setpoint)):
    for ext_ref in bldg_graph.objects(s, REF.hasExternalReference):
        modbus_addr = bldg_graph.value(ext_ref, MODBUS.address)
        if modbus_addr is not None:
            print(type(modbus_addr.value))
    
```
