```python
from pathlib import Path
import brickschema

ttl_filepath = Path.cwd().joinpath('brick.ttl')
# load in data files from your file system
g = brickschema.Graph(load_brick=True)
# Load QUDT ontologies from the web
g.load_file("http://qudt.org/schema/qudt/")        # QUDT schema
g.load_file("http://qudt.org/vocab/quantitykind")  # quantity kinds (e.g., Length, Power)
g.load_file("http://qudt.org/vocab/unit")
# load the file to validate
g.load_file(ttl_filepath)

```

```python
# validate the brick model
print("Validating model...")
valid, report_graph, result_text = g.validate()
print(f'Is the file valid? {valid}')
```
